```python
import numpy as np

import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({'font.size': 12})

x = [1, 2, 4, 8, 12]
f1 = [0.030, 0.019, 0.021, 0.032, 0.030]
f2 = [0.018, 0.019, 0.023, 0.036, 0.033]
f3 = [0.005, 0.008, 0.009, 0.013, 0.043]

plt.plot(x, f1, label='first touch', marker='o')
plt.plot(x, f2, label='last touch', marker='o')
plt.plot(x, f3, label='time decay', marker='o')

plt.legend(fontsize=14)
plt.title('Computational time')
plt.xlabel('month')
plt.ylabel('time (sec)')

plt.show()
```
