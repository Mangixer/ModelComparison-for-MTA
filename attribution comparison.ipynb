```python
import seaborn as sns
import matplotlib.pyplot as plt
import subprocess
```

```python
df_sample_submission = pd.read_csv('/content/mta_results111.csv')
lables1 = df_sample_submission[['ID','first_touch','last_touch', 'time_decay', 'markov','shapley','shao']]
lables1 = lables1.melt(id_vars=['ID'])
lables1.head(10)
```

```python
sns.set_style("whitegrid")
plt.rc('legend', fontsize=15)
fig, ax = plt.subplots(figsize=(16, 10))
sns.barplot(x='ID', y='value', hue="variable", data=lables1)
plt.show()
```
